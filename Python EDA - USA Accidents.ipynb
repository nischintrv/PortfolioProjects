### 2. Start time

```python
df.Start_Time
```

```python
indexes = df['Start_Time'][df['Start_Time'].str.len()>19].index
```

```python
df = df.drop(index=indexes)
df
#This column had different string lengths which did not allow direct dtyoe change. so first i have deleted all these rows to proceed. 
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
df.Start_Time
```

```python
df.Start_Time[0]
```

```python
hours_of_each_acc = df.Start_Time.dt.hour
```

```python
sns.distplot(hours_of_each_acc,bins = 24,kde=False, norm_hist=True)
```

1. High % of acc. occur between 6-10am(probably due morning work commute rush)

```python
dow_of_each_acc = df.Start_Time.dt.dayofweek
dow_of_each_acc
```

```python
sns.distplot(dow_of_each_acc, bins=7, kde=False, norm_hist=True)
```

Is the distribution of acc. by hour different on weekends and weekdays?

```python
sundays_hour_of_acc = df.Start_Time[df.Start_Time.dt.dayofweek == 5].dt.hour 
sns.distplot(sundays_hour_of_acc, bins=24, kde=False, norm_hist=True)
```

On Sundays, the peak occurs between 10am - 3pm, unlike weekdays

```python
mondays_hour_of_acc = df.Start_Time[df.Start_Time.dt.dayofweek == 0].dt.hour 
sns.distplot(mondays_hour_of_acc, bins=24, kde=False, norm_hist=True)
```

```python
months_of_each_acc = df.Start_Time.dt.month
sns.distplot(months_of_each_acc, bins=12, kde=False, norm_hist=True)
```

Month wise Trend of accidents reasons
- colder months with snow and icy road conditions will contribute to more acc.
